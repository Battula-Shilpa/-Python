**Matplotlib vs Seaborn** Basic Difference

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
import pandas as pd
```

```python
#Create a same data
range= np.random.RandomState(0)
x= np.linspace(0,10,500)
y= np.cumsum(range.randn(500,6),axis=0)
```

```python
#for Matplotlib
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left');
```

```python
#for seaborn
import seaborn as sns
sns.set()
#plotting
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left');
```

```python
#Histogram in seaborn
data= np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data= pd.DataFrame(data,columns=['x','y'])

for col in 'xy':
  plt.hist(data[col],alpha=0.5)
```

```python
#smooth Estimation using Kernal Density
for col in 'xy':
  sns.kdeplot(data[col], fill= True)
```

```python
#Histogram + Kernal Density Estimation
sns.distplot(data['x'],kde = True)
sns.distplot(data['y'],kde = True);
#plt.show()
```

```python
#Passing two deimensional dataset to kde plot is simple
sns.kdeplot((data));
```

```python
#joint distribution
with sns.axes_style(''):
  sns.jointplot(x='x',y='y',data=data,kind='kde')
```
