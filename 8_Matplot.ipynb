# Intro of **MATPLOT**
-  widely used Python library for creating static, animated, and interactive visualizations. It is highly customizable and supports various types of plots, such as line plots, bar charts, histograms, scatter plots, and more.

- **Pyplot** is a module in Matplotlib, a popular plotting library in Python.

- **Line Plot** -	Time series trends (stock prices, temperature)
- **Scatter Plot** -	Correlation and relationships between variables
- **Bar Chart** - Comparing discrete categories (sales, revenue)
- **Histogram	Frequency** - distribution (exam scores, ages)
- **Box Plot** -	Outliers and statistical distribution
- **Pie Chart** -	Proportions of a whole (market share)
- **Heatmap** -	Matrices, correlation maps
- **Stacked Bar Chart** -	Part-to-whole comparisons
- **Area Chart** -	Cumulative trends over time
- **3D Plot** -	Multi-variable relationships

# Bar Plot

- A bar chart or bar graph is a chart that presents categorical data with rectangular bars with heights or lengths propotional to the values that they represent.
- The bar can be plotted vertically or horizontally

```python
import numpy as np
import matplotlib.pyplot as plt

#Creating the dataset
data = {'Apples': 120,'Mangoes': 100,'Grapes': 90,'Orange': 70}
names = list(data.keys())
price = list(data.values())

fig = plt.figure(figsize=(10,5))
plt.bar(names,price, color= ['red','yellow','green','orange'])

plt.xlabel('Fruits')
plt.ylabel('Price')
plt.title('Fruits Prices')

plt.show()
```

#Box Plot

- Box plot is very helpful in viewing the summary of dataset in an efficient way also box plot helps you in doing outliers analysis

**How a Box Plot is Constructed**
- A box plot is built using five-number summary statistics:

1. **Minimum**  The smallest data point (excluding outliers).
2. **First Quartile (Q1, 25th percentile)**  The median of the lower half of the dataset.
3. **Median (Q2, 50th percentile)**  The middle value of the dataset.
4. **Third Quartile (Q3, 75th percentile)**  The median of the upper half of the dataset.
5. **Maximum**  The largest data point (excluding outliers).

**How the Box Plot is Generated:**
1. **Sort the Data**: Arrange the dataset in ascending order.
2. **Compute the Five-Number Summary:**
- Find Q1, Q2 (median), and Q3.
- Compute IQR = Q3 - Q1.
- Set the whiskers at Q1 - 1.5IQR (lower bound) and Q3 + 1.5IQR (upper bound).
3. **Identify Outliers**:
Any point below Q1 - 1.5IQR or above Q3 + 1.5IQR is marked as an outlier.
Plot the Box, Whiskers, and Outliers.

```python
data = [3,5,9,1,15,29,32]

plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#data preparation
total = [20,4,6,28,48,19,20,47,65,68]
order = [1,3,4,6,22,45,78,34,32,56]
discount = [30,10,20,15,28,48,50,60,90,80]

data = list([total,order,discount])

plt.boxplot(data,showmeans= True)
plt.title('Box Plot')
plt.grid(True)
plt.show()
```

# Viloin Plot
- Same as box plot ,instead of shape
- Summary statistics + probability density

```python
import matplotlib.pyplot as plt
%matplotlib inline

#data preparation
total = [20,4,6,28,48,19,20,47,65,68]
order = [1,3,4,6,22,45,78,34,32,56]
discount = [30,10,20,15,28,48,50,60,90,80]

data = list([total,order,discount])

plt.violinplot(data,showmeans= True, showmedians= True)
plt.title('Viloin Plot')
plt.grid(True)
plt.show()
```

# Pie Chart

- used to show percentage or propotional data,good for displaying data for around 6 categoties or fewer

```python
import matplotlib.pyplot as plt

#prepare data
labels = ['frogs','dogs','cats','rabbits']
sizes = [15,20,54,23]

#Add figure
fig= plt.figure(figsize=(10,5))
#add axes
ax1 = plt.subplot()

ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, explode=(0.1,0.1,0,0))

plt.title('Piechart')
plt.show()
```

# Area plot

```python
#Area is covered by some curve
import numpy as np
import matplotlib.pyplot as plt

x= range(1,15)
y= (1,4,6,8,4,6,2,8,2,6,4,9,4,7)

#creating a area plot
plt.stackplot(x,y,colors= 'k', alpha= 0.5)
plt.plot(x,y,color='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot')
plt.grid(True)
plt.show()
```
